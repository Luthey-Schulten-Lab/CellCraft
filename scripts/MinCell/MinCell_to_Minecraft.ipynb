```python
workDir = '/home/zane/MinCell/MinCellCME-RDME/RDME_CME_ODE/main/new_folder'
filename = workDir + '/MinCell_jLM_RDME_CME_ODE_polysomes.lm'
traj = RDMEFile(filename,replicate=1)
```

```python
np.argwhere(traj.siteLattice==0)
```

```python
site_types = {'1':'membrane', '2':'cytoplasm', '3':'ribosomes', '4':'DNA', '5':'outer_cytoplasm'} #'0':'extracellular', 
```

```python
for idx, region in site_types.items():
    print(idx, region)
    npfile = region + '.npy'
    np4nm_file = region + '_4nm.npy'
    sites = traj.siteLattice==int(idx)
    sites_xyz = np.argwhere(sites==True)
    sites_4nm = np.full((128,128,128), False)
    for site in sites_xyz:
        sites_4nm[2*site[0],2*site[1],2*site[2]] = True
        sites_4nm[2*site[0]+1,2*site[1],2*site[2]] = True
        sites_4nm[2*site[0],2*site[1]+1,2*site[2]] = True
        sites_4nm[2*site[0],2*site[1],2*site[2]+1] = True
        sites_4nm[2*site[0]+1,2*site[1]+1,2*site[2]] = True
        sites_4nm[2*site[0]+1,2*site[1],2*site[2]+1] = True
        sites_4nm[2*site[0],2*site[1]+1,2*site[2]+1] = True
        sites_4nm[2*site[0]+1,2*site[1]+1,2*site[2]+1] = True
    np.save(npfile, sites)
    np.save(np4nm_file, sites_4nm)
```
